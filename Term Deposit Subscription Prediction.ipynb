#### 1. What is the distribution of age among the clients?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("banking_data.csv")
df.head(5)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='age',bins=30, kde=True)
plt.title('Distribution of Age Among Clients')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

* Most of the clients belongs to the age group 30 to 40.

* Clients less 25 and more than 60 years of age are very less.

* Clients having blue-collar jobs are the most.

* Apart from unkown job type students are least in number.

* Client belongs to management domain are the second highest in number and are very close to blue-collar

#### Are there any correlations between different attributes and the likelihood of subscribing to a term deposit?

```python
numerical_columns = df.select_dtypes(include=['number'])
correlation_matrix = numerical_columns.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```
