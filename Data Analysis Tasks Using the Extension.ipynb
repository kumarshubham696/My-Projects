#### Usage Analysis

```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

df = pd.read_excel('askmeoffersdataset.xlsx')

# Group by E-commerce platform and calculate average rating and total saving
usage_analysis = 
df.groupby('E-commerce').agg({'Rating': 'mean', 'Total Saving': 'sum'}).reset_index()

print("Usage Analysis:")
print(usage_analysis)
```

* Evaluated how the extension impacts user behavior on different e-commerce platforms.
* Calculated the average rating and total savings for each platform

```python
import matplotlib.pyplot as plt
fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('E-commerce Platform')
ax1.set_ylabel('Average Rating', color=color)
ax1.bar(df['E-commerce'], df['Rating'].mean(), color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Total Savings (in â‚¹)', color=color)
ax2.bar(df['E-commerce'], df['Total Saving'].sum(), color=color)
ax2.tick_params(axis='y', labelcolor=color)

plt.title('Usage Analysis: Average Rating and Total Savings by E-commerce Platform')
plt.show()
```

#### Effectiveness Assessment

```python
# Filter the dataset for coupon applicable products
coupon_applicable_products = df[df['Coupon Applicable'] == 'yes']

# Calculate the redemption rate
redemption_rate = (coupon_applicable_products['Coupon Applicable'].count() / df['Coupon Applicable'].count()) * 100

print("\nEffectiveness Assessment:")
print(f"Redemption Rate: {redemption_rate:.2f}%")
```

* Analyzed the effectiveness of the coupon application feature.
* Calculated the redemption rate: 71.33% of products have coupons applied.

```python
# Data for visualization
redemption_rate = 71.33
non_redemption_rate = 100 - redemption_rate

# Plotting the pie chart
labels = ['Redemption Rate', 'Non-Redemption Rate']
sizes = [redemption_rate, non_redemption_rate]
colors = ['lightgreen', 'lightcoral']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Effectiveness Assessment: Coupon Redemption Rate')
plt.axis('equal')
plt.show()
```
