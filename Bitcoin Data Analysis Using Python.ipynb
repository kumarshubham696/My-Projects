## 1. Create a DataFrame:

#### Importing Neccessary Libariries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2. Add Data:

```python
open_prices = np.random.uniform(20000, 40000, 30)
low_prices = open_prices - np.random.uniform(0, 1000, 30)
high_prices = open_prices + np.random.uniform(0, 1000, 30)
close_prices = [np.random.uniform(low, high) for low, high in zip(low_prices, high_prices)]
volume = np.random.uniform(1000000, 5000000, 30)
market_cap = np.random.uniform(1e9, 2e9, 30)
```

```python
data = {
    'Date': pd.date_range(start='2024-08-08', periods=30, freq='D'),
    'Open': open_prices,
    'High': high_prices,
    'Low': low_prices,
    'Close': close_prices,
    'Volume': volume,
    'MarketCap': market_cap
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

## 5. For Loop and Conditional Statements:

#### Use a for loop to iterate over the rows of the DataFrame and calculate the daily price change (difference between `Close` and `Open`), adding this as a new column `Price Change`.

```python
price_change = []
for index, row in df.iterrows():
    change = row['Close'] - row['Open']
    price_change.append(change)
```

```python
df['PriceChange'] = price_change
```

```python
df
```

#### Apply an if-else condition to create another column `Price Trend`, where:

##### If the `Close` price is higher than the `Open` price, set the value to "Up."

##### Otherwise, set the value to "Down."

```python
def trend(row):
    if row['Close'] > row['Open']:
        return 'Up'
    else:
        return 'Down'
```

```python
df['PriceTrend'] = df.apply(trend, axis = 1)
```
