```python
# Data Description
# This is real investment data taken from crunchbase.com
import pandas as pd
```

```python
# 1. Reading the Rounds2, companies.txt files.
rounds2 = pd.read_excel("D:\\Received Files\\rounds2.xlsx")
rounds2
```

```python
companies = pd.read_csv("D:\\ITP Mini Project\\Project 2-2-1\\Project 2\\Dataset\\companies.txt", sep='\t')
companies
```

```python
# 2. How many unique companies are present in rounds2?

unique_companies_rounds2 = rounds2['company_permalink'].nunique()
unique_companies_rounds2
```

```python
# 3. How many unique companies are present in the companies file?

unique_companies_companies = companies['permalink'].nunique()
unique_companies_companies
```

```python
# 4.Write the code to convert companies[‘permalink’] and Round2[‘company_permalink’] columns
# to uppercase.

companies['permalink'] = companies['permalink'].str.upper()
```

```python
companies['permalink']
```

```python
rounds2['company_permalink'] = rounds2['company_permalink'].str.upper()
rounds2['company_permalink']
```

```python
# 5. Are there any companies in the rounds2 file which are not present in companies.txt ?

companies_not_in_rounds2 = rounds2[~rounds2['company_permalink'].isin(companies['permalink'])]
companies_not_in_rounds2
```

```python
# 6. Merge the two data frames so that all variables (columns) in the company's frame are added
# to the rounds2 data frame. Name the merged frame master_dataframe. How many observations
# are present in master_frame ?

master_dataframe = pd.merge(rounds2, companies, how='left', left_on='company_permalink', right_on='permalink')
observations_count = master_dataframe.shape[0]
master_dataframe
```

```python
observations_count
```

```python
# 7. Write the code to drop the redundant column ‘'company_permalink' from the
# master_dataframe.

master_dataframe.drop('company_permalink', axis=1, inplace=True)
master_dataframe
```

```python
# 8. Write the code to find the percentage missing value (column-wise) in master_dataframe. Drop
# unnecessary columns.

missing_values_percentage = master_dataframe.isnull().mean() * 100
master_dataframe.dropna(axis=1, thresh=len(master_dataframe) * 0.7, inplace=True)
missing_values_percentage
```

```python
# 9. Create a pivot table to compare the mean and median values for ‘raised_amount_usd’ and
# ‘funding_round_type’ across different funding round categories..

pivot_table = master_dataframe.pivot_table(values='raised_amount_usd', index='funding_round_type', aggfunc=['mean', 'median'])
pivot_table
```

```python
# 10. Filter the master_dataframe on ‘raised_amount_usd’ such as ‘raised_amount_usd’ lie
# between 4 Millions to 15 Millions.

filtered_master_dataframe = master_dataframe[(master_dataframe['raised_amount_usd'] >= 4000000) & (master_dataframe['raised_amount_usd'] <= 15000000)]
filtered_master_dataframe
```

```python
# 11.Identify the top 9 countries in terms of highest investment in ‘venture’ fund type.

top_9_countries_venture = master_dataframe[master_dataframe['funding_round_type'] == 'venture'].groupby('country_code')['raised_amount_usd'].sum().nlargest(9)
top_9_countries_venture
```

```python
# 12. Create the ‘main_category’ column by extracting the main sector using the column
# ‘category_list’.
# Hint - Use the Lambda function or string function

master_dataframe['main_category'] = master_dataframe['category_list'].apply(lambda x: x.split('|')[0] if pd.notnull(x) else x)
master_dataframe['main_category']
```

```python
# 13. Using the user defined function convert the ‘rasied_amount_used’ column into a categorical
# column as followa) If the amount is less than 5 Millions then recode as TypeA.
# b) If the amount is greater than 5 Millions and less than 7 Millions then recode as TypeB.
# c) If the amount is greater than 7 Millions then recode as TypeC

def categorize_amount(amount):
    if amount < 5000000:
        return 'TypeA'
    elif 5000000 <= amount < 7000000:
        return 'TypeB'
    else:
        return 'TypeC'

master_dataframe['raised_amount_category'] = master_dataframe['raised_amount_usd'].apply(categorize_amount)
```
